```
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```
# Tanımla
x = np.linspace(0, 5, 21)
y = x ** 3
```

```
# Çizdir
plt.plot(x, y, 'r')
plt.xlabel('X Ekseni')
plt.ylabel('Y Ekseni')
plt.title('Başlık')
plt.show()
```

```
# Çoklu Çizim
plt.subplot(1,2,1)
plt.plot(x,y,'r--')
plt.subplot(1,2,2)
plt.plot(y,x,'g*-')
```

```
plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(y,x,'g*-')
plt.subplot(2,2,3)
plt.plot(x,y,'r--')
plt.subplot(2,2,4)
plt.plot(y,x,'g*-')
```

```
fig = plt.figure()

axes = fig.add_axes([5, 0, 1.5, 2.5])

axes.plot(x, y, 'r')
axes.set_xlabel('X Ekseni')
axes.set_ylabel('Y Ekseni')
axes.set_title('Başlık')
```

```
fig = plt.figure()

axesA = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axesB = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axesA.plot(x,y, 'y')
axesA.set_xlabel('x Ekseni A')
axesA.set_ylabel('y Ekseni A')
axesA.set_title('Başlık A')

axesB.plot(y, x, 'r')
axesB.set_xlabel('x Ekseni B')
axesB.set_ylabel('y Ekseni B')
axesB.set_title('Başlık B')
```
